```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
from json import JSONEncoder
from transformers import AutoModelForSequenceClassification
from transformers import TFAutoModelForSequenceClassification
from transformers import AutoTokenizer, AutoConfig
from scipy.special import softmax
from tqdm import tqdm
```

```python
hashtag = "FIFAWorldCupQatar2022"
```

```python
session = requests.Session()
headers = {}
headers['Authorization'] = "Bearer YOURTOKENHERE"
session.headers = headers
response = session.get("https://api.twitter.com/2/tweets/search/recent?query=%23"+hashtag+"%20lang%3Aen&max_results=100")
```

```python
stringResponse = json.loads(response.text)
```

```python
data = stringResponse['data']
for i in tqdm(range(50)):
    if('next_token' in stringResponse['meta']):
        response = session.get("https://api.twitter.com/2/tweets/search/recent?query=%23"+hashtag+"%20lang%3Aen&max_results=100&next_token="+stringResponse["meta"]["next_token"])
        stringResponse = json.loads(response.text)
        data.extend(stringResponse['data'])
```

```python
#Removing spams
spamlessData = []
for tweet in data:
    spamlessData.append(tweet['text'])
print(len(spamlessData))
spamlessData = list(set(spamlessData))
print(len(spamlessData))

data = spamlessData
```

```python
tweetWords = ''
stopwords = set(STOPWORDS)
stopwords = list(stopwords)
stopwords.append(hashtag.lower())
stopwords.extend(['t', 'co', 'https', 'http', 'rt', '3dx', '3dcg', 'ipsgqg0nbe', 'now', 'given', 'want'])
for tweet in data:
    tokens = tweet.split()
    
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
    
    tweetWords += " ".join(tokens) + " "
```

### Word cloud to see what are people talking about when they use this hashtag

```python
wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(tweetWords)

plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```

```python
hashtag_dict = {}
for tweet in data:
    tokens = tweet.split()
    
    for i in range(len(tokens)):
        if(tokens[i][0] == '#'):
            if(tokens[i].lower() in hashtag_dict):
                hashtag_dict[tokens[i].lower()] += 1
            else:
                hashtag_dict[tokens[i].lower()] = 1
hastag_dict = pd.DataFrame(hashtag_dict.items(), columns=['hashtag', 'frequency'])
```

### Which hashtags is this tag most coupled with

```python
hastag_dict.sort_values(by=['frequency'], ascending=False).reset_index(drop=True)[1:11]
```

```python
# Preprocess text (username and link placeholders)
def preprocess(text):
    new_text = []
    for t in text.split(" "):
        t = '@user' if t.startswith('@') and len(t) > 1 else t
        t = 'http' if t.startswith('http') else t
        new_text.append(t)
    return " ".join(new_text)
MODEL = f"DownloadPathToModel"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
config = AutoConfig.from_pretrained(MODEL)
# PT
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
#model.save_pretrained(MODEL)
```

### Using Pretrained RoBERTa to see how positive or negative these tweets on this hashtag are

```python
tweetSentimentList = []
sentimentAverage = {"Positive": 0.0, "Negative": 0.0, "Neutral": 0.0}
for tweet in tqdm(data):
    text = tweet
    text = preprocess(text)
    encoded_input = tokenizer(text, return_tensors='pt')
    output = model(**encoded_input)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    ranking = np.argsort(scores)
    ranking = ranking[::-1]
    dic = {}
    for i in range(scores.shape[0]):
        l = config.id2label[ranking[i]]
        s = scores[ranking[i]]
        dic[l] = float(s)
        sentimentAverage[l] += float(s)
    
    tweetSentimentList.append([tweet, dic])
    
sentimentAverage['Positive'] = sentimentAverage['Positive'] / len(data)
sentimentAverage['Negative'] = sentimentAverage['Negative'] / len(data)
sentimentAverage['Neutral'] = sentimentAverage['Neutral'] / len(data)
```

#### Below we can see that 50% neutral tweets about this topic are just news tweets

```python
arr = [sentimentAverage['Positive'], sentimentAverage['Negative'], 1-sentimentAverage['Negative']-sentimentAverage['Positive']]

lab = ["😄", "😡", "😶"]
colors = ["#32CD32", "#FF4500", "#FFE4E1"]
plt.pie(arr, labels = lab, colors = colors, autopct='%.2f', radius=2, textprops={'fontsize': 20})
plt.show() 
```

```python
tweetSentimentDf = []
for tweet in tweetSentimentList:
    tweetSentimentDf.append([tweet[0], tweet[1]['Positive'], tweet[1]['Negative'], tweet[1]['Neutral']])
    
tweetSentimentDf = pd.DataFrame(tweetSentimentDf, columns = ['tweet', 'Positive', 'Negative', 'Neutral'])
```
